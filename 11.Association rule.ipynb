The Objective of this assignment is to introduce students to rule mining techniques, particularly focusing on market basket analysis and provide hands on experience.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
pip install mlxtend
```

```python
#1.Load Dataset
df=pd.read_excel("Online retail.xlsx",header=None,names=["Items"])
```

```python
#2.Preprocessing 

#Remove rows with missing or empty values
df=df.dropna(subset=["Items"])

#Remove duplicates
df=df.drop_duplicates()


#Convert comma-seperated string to Python list
df["Items"]=df["Items"].apply(lambda x: str(x).strip().split(","))

```

```python
#3.Transaction Encoding (One-Hot)
te=TransactionEncoder()
te_array=te.fit(df["Items"]).transform(df["Items"])
basket=pd.DataFrame(te_array, columns=te.columns_)
```

```python
#4.Apply Apriori Algorithm
#min_support=0.1 means items must appear in at least 10% of all transactions
freq_items=apriori(basket,min_support=0.1, use_colnames=True)

print("\nFrequent Itemset:\n")
print(freq_items)
```

```python
#5.Generate Association Rules

rules=association_rules(freq_items, metric="confidence",min_threshold=0.5)

#Sort rules by lift(most meaningful at top)
rules=rules.sort_values(by="lift",ascending=False)

print("\nAssociation Rules:\n")
print(rules)
```

```python
#Optional:Display only the best rules

print("\nTop 10 Strongest Rules:\n")
print(rules.head(10))
```

## Interview Questions

1. What is Lift and why is it important in Association Rules?

Lift is a measure that tells us how strongly two items are associated compared to their independent occurrence. It shows whether the presence of item A actually increases the likelihood of item B being purchased. A lift value greater than 1 indicates a positive relationship (A and B occur together more than expected by chance), a value equal to 1 indicates no relationship, and a value less than 1 indicates a negative relationship. Lift is important because it identifies truly meaningful and useful rules, helping us avoid misleading patterns that might appear strong due to high confidence but actually happen just because an item is naturally popular.

##3. What are some limitations or challenges of Association Rule Mining?

Association rule mining has several limitations. It often generates a very large number of rules, many of which may not be useful, requiring additional filtering. The Apriori algorithm is computationally expensive, especially for large datasets, because it repeatedly scans the database. The technique also identifies correlations but not causation, so rules must be interpreted carefully. Rare but important items might be ignored if the support threshold is set too high. Additionally, association rules work best with categorical data, so continuous data must be discretized before use, which may lead to information loss. Finally, interpreting rules requires domain knowledge to understand why items may appear together.
